```python
import pandas_datareader.data as web
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = web.DataReader("AAPl","yahoo", start= "2015-9-1", end= "2019-12-31")
```

```python
df
```

```python
P = df['Adj Close']
```

```python
P.head()
```

```python
P.plot()
```

```python
P.diff()        # diferences between days
```

```python
r = P.diff()/P         #cal Rate of return 
r.plot()
```

```python
r[-100:].plot()        # from 100 days to now
```

```python
P.rolling(window=20).mean()  # with rolling(numbers of window) to cal Moving average(20days)
```

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=100).mean().plot()
```
